```python
from diffusers import StableDiffusionPipeline
import torch
import matplotlib.pyplot as plt
```

```python
! pip show torch
```

```python
model_01="dreamlike-art/dreamlike-diffusion-1.0"
model_02="stabilityai/stable-diffusion-xl-base-1.0"

pipe=StableDiffusionPipeline.from_pretrained(model_01,torch_dtype=torch.float16)
pipe=pipe.to("cuda")
```

```python
prompt="dreamlike art a cute woman with laptop studying in vit bhopal university and solving dsa problems on her laptop happy university girl, ethereal lighting, academic setting"
```

```python
my_img=pipe(prompt).images[0]
plt.imshow(my_img)
plt.axis('off')
```

```python
prompt_2="A majestic, hyper-detailed portrait of a celestial fox, rendered in an intricate blend of volumetric lighting and digital painting. The fox's fur is composed of a nebula of iridescent colors: deep sapphire, vibrant magenta, and glittering gold. It stands atop a crystal mountain range under a twilight sky filled with twin moons and swirling auroras. Focus on dynamic composition, ethereal glow, and extreme fine detail. Art styles: trending on ArtStation, fantasy illustration, octane render, unreal engine, photorealistic details."
my_img=pipe(prompt_2).images[0]
plt.imshow(my_img)
plt.axis('off')
```
