**PROMPT TEMPLATES**

```python
from langchain_core.prompts import ChatPromptTemplate
system_template="tell me the recipe of {dish}"
prompt_template=ChatPromptTemplate.from_messages([("system",system_template),("user","{dish}")])
```

```python
prompt=prompt_template.invoke({"dish":" sandwiches"})

prompt
```

```python
response =model.invoke(prompt)
print(response.content)
```

**Templates**

```python
from langchain_core.prompts import PromptTemplate
recipe_template=PromptTemplate.from_template("tell me the recipe of {dish}")
recipe_prompt=recipe_template.invoke({"dish": "omlette"})
response=model.invoke(recipe_prompt)
print(response.content)
```

**CHAINS , RUNNABLES AND TASKS**
[DOCUMENTATIONS](https://docs.langchain.com/oss/javascript/migrate/langchain-v1#runnables)

```python
from langchain_core.output_parsers import StrOutputParser
recipe_chain=recipe_template|model|StrOutputParser()
print(recipe_chain.invoke({"dish":"noodles"}))
```

**CASCADING CHAINS**
- specify Vegeterian or Non Vegeterian

```python
from langchain_core.prompts import PromptTemplate
veg_propmt_template = PromptTemplate.from_template("Is the following recipe vegetarian or non-vegetarian? Recipe: {recipe_content}\nAnswer with 'Vegetarian', 'Non-Vegetarian', or 'Both possible'.")
veg_response=model.invoke(veg_propmt_template.invoke({"recipe_content": "Example recipe"}))
print(veg_response.content)
```

```python
from langchain_core.runnables import RunnableSequence, RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser

composed_chain = (
    {"recipe_content": recipe_chain}
    | veg_propmt_template
    | model
    | StrOutputParser()
)
```

```python
composed_chain.invoke("mutton kebab")
```

**RUNNABLE LAMBDAS**

```python
from langchain_core.runnables import RunnableLambda
print_veg_nonveg=RunnableLambda(lambda x: print(f"veg or non veg: {x}"))
new_chain=composed_chain|print_veg_nonveg
response=new_chain.invoke("rameon")
print(response)
```
